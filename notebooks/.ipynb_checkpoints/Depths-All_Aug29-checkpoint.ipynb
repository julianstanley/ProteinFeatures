```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
depths = pd.read_csv("../depths_out_20190829_152504.csv")
```

```python
depths["AA-Protein"] = [f'{aa}:{protein}' for aa,protein in zip(depths["AA"], depths["protein"])]
depths["AA-Position"] = [f'{aa}:{position}' for aa,position in zip(depths["AA"], depths["position"])]
depths["AA-Position-Protein"] = [f'{aa}:{position}:{protein}' for aa,position,protein in zip(depths["AA"], depths["position"], depths["protein"])]
```

```python
depths.head()
```

## What is the minimum depth for each amino acid?

Answer: 1.0 for all amino acids except for F and H. (actually closer to 0.999999, but rounded to 2 decimal places) 

```python
amino_acids = depths["AA"].unique()
min_depths = {}
for aa in amino_acids:
    depths_sorted = sorted([round(float(depth), 3) for depth in depths[depths["AA"] == aa]["depth"]])
    min_depths[aa] = depths_sorted[0]
```

```python
print(min_depths)
```

## How much variability in depth do we see per protein and per aa?

Answer: It depends...but we can look at some histograms to get an idea.


```python
depths_variation = {}
for aa_pos_prot in depths["AA-Position-Protein"].unique():
    these_depths = depths[depths["AA-Position-Protein"] == aa_pos_prot]['depth']
    depth_var = np.std(these_depths)
    
    protein = aa_pos_prot.split(':')[2]
    if protein in depths_variation:
        depths_variation[protein].append(depth_var)
    else:
        depths_variation[protein] = [depth_var]
    
```

```python
for protein, variation in depths_variation.items():
    print(f'Protein: {protein}: standard dev of depths: {[round(float(x),2) for x in variation[0:5]]} [...]')
```

```python
fig, axes = plt.subplots(nrows=7, ncols=1, figsize=(10,40))

for plot, protein in zip(range(7), depths_variation.keys()):
    axes[plot].hist(depths_variation[protein], range = [0,2])
    axes[plot].set_title(protein)
```

```python
depths_variation_aa = {}
for aa_pos_prot in depths["AA-Position-Protein"].unique():
    these_depths = depths[depths["AA-Position-Protein"] == aa_pos_prot]['depth']
    depth_var = np.std(these_depths)
    
    aa = aa_pos_prot.split(':')[0]
    if aa in depths_variation_aa:
        depths_variation_aa[aa].append(depth_var)
    else:
        depths_variation_aa[aa] = [depth_var]
```

```python
for aa, variation in depths_variation_aa.items():
    print(f'AA: {aa}: standard dev of depths: {[round(float(x),2) for x in variation[0:5]]} [...]')
```

```python
fig, axes = plt.subplots(nrows=20, ncols=1, figsize=(10,100))

for plot, aa in zip(range(21), depths_variation_aa.keys()):
    axes[plot].hist(depths_variation_aa[aa], range = [0,2])
    axes[plot].set_title(aa)
```

```python
atom_to_depth = {}
for protein,aa,position,depth in zip(depths["protein"], depths["AA"], depths["position"], depths["depth"]):
    name = f'{protein}_{aa}_{position}'
    if name in atom_to_depth:
        atom_to_depth[name].append(depth)
    else:
        atom_to_depth[name] = [depth]
```

```python
same = 0
different = 0
for name, depths in atom_to_depth.items():
    if len(set(depths)) > 1:
        different += 1
        print(f'for {name} depth was off by a stdev of {round(np.std(depths),2)}')
    else:
        same += 1
        
print(f'{same} atoms had the same depth 5 times, whereas {different} atoms did not.')
```
